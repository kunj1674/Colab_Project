**1. Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.**

```python
# Method one.
actor_list=[]
for i in eval(df[df['profit']==df['profit'].max()][['cast']].values[0][0]):
  actor_list.append(i['name'])

print(actor_list)
```

```python
# Method two
actor_list1=[]
for i in df[df['profit']==df['profit'].max()]['cast']:
  for j in eval(i):
    actor_list1.append(j['name'])


print(actor_list1)
```

```python
#Director
director=[]
for i in eval(df[df['profit']==df['profit'].max()][['crew']].values[0][0]):
  if i['job']=='Director':
    director.append(i['name'])

print(director)
```

```python
#Producer
Producer=[]
for i in eval(df[df['profit']==df['profit'].max()][['crew']].values[0][0]):
  if i['job']=='Producer':
    Producer.append(i['name'])

print(Producer)
```
