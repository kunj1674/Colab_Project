Data Visualization using the library named Metplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path= '/content/drive/MyDrive/Colab Notebooks/Copy of pp-2018.csv'
```

```python
df = pd.read_csv(path)
df.head(7)
```

```python
df = pd.read_csv(path, names= ["TID","Price","Sale Date", "Postcode",
                                    "Property Type","New Build","Tenure",
                                    "PAON","SAON","Street","Locality",
                                    "Town/City","District","County",
                                    "PPD Category","Record Status"],
                            parse_dates= ["Sale Date"])
```

```python
df.head(7)
```

```python
df.info()
```

* Next we will filter the data to get all the transaction for a particular year and month ( say July 2018 for example). We will only keep transactions of type "A" (Additions) in the data frame and get rid of Changes and Deletions. And for the sake of simplicity we will also remove transactions for property type "O" (Other) and focus only on Detached ("D"), Semi Detached ("S"), Flats ("F") and Terraced ("T") houses.

```python
df.shape
```

```python
df['Property Type'].value_counts()
```

```python
df = df[df['Property Type'] != 'O']
df = df[df['Record Status'] == 'A']
df = df[(df['Sale Date'].dt.year == 2018)]
```

```python
df.shape
```

First objective : Obtain the trend of sales volume across months

```python
df.columns
```

```python
df['month']= df['Sale Date'].dt.month
df['day']=df['Sale Date'].dt.day
```

```python
df.head(7)
```

```python
Monthly_Sales=df['month'].value_counts().sort_index()
Monthly_Sales
```

```python
plt.figure(figsize=(9,3))
Monthly_Sales.plot(kind='bar')
plt.title("Total Sales Volume of each month")
plt.xlabel('Month')
plt.ylabel('Sales Volume')
```

```python
plt.figure(figsize=(9,3))
Monthly_Sales.plot(kind='line')
plt.title("Total Sales Volume of each month")
plt.xlabel('Month')
plt.ylabel('Sales Volume')
```

```python
plt.figure(figsize=(9,3))
Monthly_Sales.plot(kind='pie')
plt.title("Total Sales Volume of each month")
plt.xlabel('Month')
plt.ylabel('Sales Volume')
```

Second objective : Observe the daily sales volume in the month of July 2018

```python
july_sales = df[df['month']== 7]
```

```python
july_sales['day'].value_counts().sort_index()
```

```python
july_sales = df[df['month']== 7]['day'].value_counts().sort_index()

july_sales.plot(kind='bar')
plt.title("Total Sales Volume in Month of July")
plt.xlabel("dates")
plt.ylabel("Sales Volume")
```

```python
july_sales = df[df['month']== 7]['day'].value_counts().sort_index()

plt.figure(figsize=(9,3))
july_sales.plot(kind='line')
plt.title("Total Sales Volume in Month of July")
plt.xlabel("dates")
plt.ylabel("Sales Volume")
```

Third Objective : Comparing the daily sales volume of different property types in July

```python
july_sales = df[df['month']==7]
```

```python
df.columns
```

```python
day_Sales=july_sales.groupby(['day','Property Type'])['day'].value_counts().sort_index().unstack()
day_Sales
```

```python
plt.figure(figsize=(9,3))
day_Sales.plot(kind='line')
plt.title("Sales Volume of the different property type in july")
plt.xlabel("dates")
plt.ylabel("Sales Volume")
```

```python
plt.figure(figsize=(9,3))
day_Sales.plot(kind='bar',stacked=True)
plt.title("Sales Volume of the different property type in july")
plt.xlabel("dates")
plt.ylabel("Sales Volume")
```

```python
plt.figure(figsize=(9,3))
day_Sales.plot(kind='bar')
plt.title("Sales Volume of the different property type in july")
plt.xlabel("dates")
plt.ylabel("Sales Volume")
```

Fourth Objective : Draw comparison among different property types in terms of sales

```python
category_sales = df['Property Type'].value_counts()
category_sales
```

```python
plt.figure(figsize=(9,3))
category_sales.plot(kind="bar")

plt.title("Comparision about different property type in terms of sales")
plt.xlabel("Property Type")
plt.ylabel("Sales Volume")
```

```python
plt.figure(figsize=(9,3))
category_sales.plot(kind="pie")

plt.title("Comparision about different property type in terms of sales")
```

Fifth Objective : Average house prices for different property types in July

```python
df.columns
```

```python
july_price = july_sales.groupby(['Property Type'])['Price'].mean()
july_price
```

```python
plt.figure(figsize=(9,3))
july_price.plot(kind='bar')

plt.title("Average Price for Different Property Type in July")
plt.xlabel("Property Type")
plt.ylabel("Average Price")
```
